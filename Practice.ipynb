```python
#creating request and accessing tags




from bs4 import BeautifulSoup
import requests
from fake_useragent import UserAgent
ua= UserAgent()
header= {"user-agent": ua.chrome}
response= requests.get("https://www.google.com/gmail/", headers= header)
print(response.content)
soup= BeautifulSoup(response.content, "lxml")
soup.prettify()
title= soup.title
print (title.get("style"))
title.string.replace_with("Hello")
print(title.string)
```

```python
#navigating parse tree




print(title.descendents)        #navigating all children
print(title.parent)             #navigating immediate parent
print(title.parents)            #navigating all parents
print(title.next_sibing)        #navigating next sibling
print(title.previous_sibling)    #navigating previous sibling
```

```python
#regular expression



import re
regex= re.compile("[a-zA-Z]")       #returns regular expression object
print(regex.match("akshay"))

reg= re.compile("abc")
print(reg.match("ab"))               #using match function to match regex in string
print(reg.match("abcdefgh"))

reg1= re.compile("[ab]*")          #using * for 0 or more occurence of [ab]
print(reg1.match(""))

reg2= re.compile("[ab]+")         #using + for 1 or more occurence of [ab] 
print(reg2.match(""))

reg3= re.compile("a?c")           #using ? for 0 or 1 occurence of a 
print(reg3.match("acdef"))
print(reg3.match("aabcc"))

reg4= re.compile("^a[a-z]")      #using metacharacter ^
print(reg4.match("abe"))
print(reg4.match("bce"))

reg5= re.compile("a|b")          #using metacharacter |
print(reg5.match("bdefbc"))

reg6= re.compile("abc$")         #using metacharacter $
print(reg6.match("abc"))


```
